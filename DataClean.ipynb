```python
#plot for top 10 songs
top_10_songs = ds.nlargest(10,'Spotify Popularity')


plt.figure(figsize=(8, 8))
plt.barh(top_10_songs['Album Name'], top_10_songs['Spotify Popularity'], color='red')
plt.xlabel('Spotify Popularity',fontweight='bold')
plt.ylabel('Track Name',fontweight='bold')
plt.title('Top 10 Most-Streamed Spotify Songs')
plt.gca().invert_yaxis()  # Highest popularity at the top
plt.show()
```

```python

# Ensure 'Spotify Popularity' is numeric
ds['Spotify Popularity'] = pd.to_numeric(ds['Spotify Popularity'], errors='coerce')

# Group by artist and calculate the average popularity
artist_popularity = ds.groupby('Artist')['Spotify Popularity'].mean()

# Sort artists by popularity
most_popular_artists = artist_popularity.nlargest(10)

# Create a horizontal bar chart
plt.figure(figsize=(12, 8))
plt.barh(most_popular_artists.index, most_popular_artists.values, color='orange')
plt.xlabel('Average Spotify Popularity',fontweight='bold')
plt.ylabel('Artist Name',fontweight='bold')
plt.title('Top 10 Most Popular Artists')
plt.gca().invert_yaxis()  # Highest popularity at the top
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SelectFromModel
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
iris_data = pd.read_csv(url, names=column_names)

# Convert species to numerical values for regression purposes
iris_data['species'] = iris_data['species'].astype('category').cat.codes

X = iris_data.iloc[:, :-1].values  # Features
y = iris_data['species'].values      # Target variable

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

lr = LinearRegression()
lr.fit(X_train, y_train)

# Perform feature selection using SelectFromModel
selector = SelectFromModel(lr, threshold='mean')  # Select features above the mean importance
selector.fit(X_train, y_train)

# Transform the training set to get selected features
X_train_selected = selector.transform(X_train)

# Get the selected feature indices
selected_features = selector.get_support(indices=True)
print("Selected features indices:", selected_features)
print("Selected features names:", [column_names[i] for i in selected_features])
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt


lr = LinearRegression()

# Perform RFE
n_features_to_select = 3  # Choose the number of features to select
rfe = RFE(estimator=lr, n_features_to_select=n_features_to_select)
rfe.fit(X_train, y_train)

# Get the selected feature indices
selected_features = rfe.get_support(indices=True)
print("Selected features indices by RFE:", selected_features)
print("Selected features names:", [column_names[i] for i in selected_features])
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

#train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create and fit the Extra Trees Classifier
etc = ExtraTreesClassifier(n_estimators=100, random_state=42)
etc.fit(X_train, y_train)

y_pred = etc.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy of Extra Trees Classifier:", accuracy)
```
